```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report

# Assuming you have already defined X and y

# Assuming 'Text' is present in the original DataFrame df
X_train, X_test, y_train, y_test = train_test_split(df[['Text', 'User', 'Platform']], df['Mood'], test_size=0.2, random_state=42)

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train['Text'])
X_test_tfidf = tfidf_vectorizer.transform(X_test['Text'])

# Convert sparse matrices to dense matrices
X_train_tfidf_dense = X_train_tfidf.toarray()
X_test_tfidf_dense = X_test_tfidf.toarray()


model_pipeline = [
    LogisticRegression(solver='liblinear'),
    SVC(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GaussianNB()
]

model_list = ['Logistic Regression', 'SVM', 'KNN', 'Decision Tree', 'Random Forest', 'Naive Bayes']
acc_list = []
cm_list = []
c_report = []

for model, model_name in zip(model_pipeline, model_list):
    model.fit(X_train_tfidf_dense, y_train)
    y_pred = model.predict(X_test_tfidf_dense)
    acc_list.append(metrics.accuracy_score(y_test, y_pred))
    cm_list.append(confusion_matrix(y_test, y_pred))
    c_report.append(f'{model_name}:\n{classification_report(y_test, y_pred)}')

c_report
```

# **Confusion Matrix**

```python
fig = plt.figure(figsize=(18,10))
for i in range(len(cm_list)):
    cm = cm_list[i]
    model = model_list[i]
    sub = fig.add_subplot(2,3,i+1).set_title(model)
    cm_plot = sns.heatmap(cm,annot=True,cmap="Blues_r")
    cm_plot.set_xlabel('predicted values')
    cm_plot.set_ylabel('actual values')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Assuming X_train_tfidf_dense and X_test_tfidf_dense are already created

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf_dense, y_train, test_size=0.2, random_state=42)

# Create a pipeline with preprocessing steps
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Optional: Standardize features
    ('classifier', RandomForestClassifier())  # You can try a different classifier
])

# Define hyperparameters to tune
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred = best_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Get the best hyperparameters
best_params = grid_search.best_params_
print(f"Best Hyperparameters: {best_params}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics


# Logistic Regression
logistic_regression = LogisticRegression(solver='liblinear')
logistic_regression.fit(X_train, y_train)
y_pred_lr = logistic_regression.predict(X_test)
accuracy_lr = metrics.accuracy_score(y_test, y_pred_lr)

# SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
accuracy_svm = metrics.accuracy_score(y_test, y_pred_svm)

# K-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
accuracy_knn = metrics.accuracy_score(y_test, y_pred_knn)

# Decision Tree
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
y_pred_dt = decision_tree_model.predict(X_test)
accuracy_dt = metrics.accuracy_score(y_test, y_pred_dt)

# Random Forest
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)
y_pred_rf = random_forest_model.predict(X_test)
accuracy_rf = metrics.accuracy_score(y_test, y_pred_rf)

# Create a pipeline with preprocessing steps
pipeline_gb = Pipeline([
    ('scaler', StandardScaler()),  # Optional: Standardize features
    ('classifier', GradientBoostingClassifier())  # Different classifier
])

# Define hyperparameters to tune
param_grid_gb = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__learning_rate': [0.01, 0.1, 0.2],
    'classifier__max_depth': [3, 4, 5]
}

# Use GridSearchCV for hyperparameter tuning
grid_search_gb = GridSearchCV(pipeline_gb, param_grid_gb, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_gb.fit(X_train, y_train)

# Get the best model
best_model_gb = grid_search_gb.best_estimator_
# Make predictions on the test set
y_pred_gb = best_model_gb.predict(X_test)
# Calculate accuracy
accuracy_gb = accuracy_score(y_test, y_pred_gb)

# Get the best hyperparameters
best_params_gb = grid_search_gb.best_params_


# Print accuracies
print(f"Gradient Boosting Accuracy: {accuracy_gb:.4f}")
print(f"Best Hyperparameters (Gradient Boosting): {best_params_gb}")
print("Logistic Regression Accuracy:", accuracy_lr)
print("SVM Accuracy:", accuracy_svm)
print("K-Nearest Neighbors Accuracy:", accuracy_knn)
print("Decision Tree Accuracy:", accuracy_dt)
print("Random Forest Accuracy:", accuracy_rf)
```
